```python
#W6A1
def wipeout(lst):
    dct = dict()
    arr = []
    for i in lst: 
        if dct.get(i,0) == 0: 
            arr.append(i)
            dct[i] = 1
    return arr

a = [int(num) for num in input().split()]
print(wipeout(a))
```

```python
#9
def adding(D1, D2):
    for item in D1.keys():
        if item not in D2:
            D2[item] = D1.get(item) 
        else: 
            D2[item] += D1.get(item)
    return D2
d1 = eval(input()) # ham eval
d2 = eval(input())

newD = {}
adding(d1,newD)
adding(d2,newD)
print(newD)
```

```python
#10
lst = [int(num) for num in input().split()]
k = int(input())
dct = {}
for i in lst: 
    dct[i] = dct.get(i,0) + 1
arr = []
for i in lst:
    if dct[k - i] != 0 and i != k - i : 
        arr.append((i,k-i))
        dct[i] -= 1
    elif i == k - i and dct[i] >= 2 and (i,i) not in arr: 
        arr.append((i,i))
        dct[i] -= 2
arr.sort()
print(arr)
```

```python
#12

lst = [int(n) for n in input().split()]
dct = {}
for i in lst: 
    dct[i] = dct.get(i,0) + 1
res = 0
rec = 0
for i in lst: 
    if dct[i] > rec: 
        res = i
        rec = dct[i]
    elif dct[i] == rec and i < res:
        res = i 
print(res)
```
