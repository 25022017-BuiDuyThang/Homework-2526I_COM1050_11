```python
#W6A1
def wipeout(lst):
    dct = dict()
    arr = []
    for i in lst: 
        if dct.get(i,0) == 0: 
            arr.append(i)
            dct[i] = 1
    return arr

a = [int(num) for num in input().split()]
print(wipeout(a))
```

```python
#2 
lst = [int(n) for n in input().split()]
arr = [0] * len(lst)
arr[0] = lst[0]
for i in range(1,len(lst)): 
    arr[i] = arr[i-1] + lst[i]
print(arr)
```

```python
#5
lst = [int(num) for num in input().split()]
dct = {"positives":0, "negatives":0, "zero":0}
for i in lst:
    if i > 0:
        dct["positives"] += 1
    elif i < 0: 
        dct["negatives"] += 1
    else:
        dct["zero"] += 1
print(dct)
```
