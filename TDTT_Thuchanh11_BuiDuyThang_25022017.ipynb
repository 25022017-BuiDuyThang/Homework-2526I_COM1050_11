```python
#3 
# KEY: use set nếu có phần tử lặp lại --> chuỗi đó lặp vô hạn không bao giờ ra 1 được 

def sum_of_power(n):
    sum = 0
    while n > 0: 
        tem = n % 10 
        sum += (tem*tem)
        n //= 10
    return sum
try: 
    n = int(input())
    s = set()
    k = n
    check = False
    while True: 
        temp = sum_of_power(k)
        if temp == 1:
            check = True
            break 
        if temp not in s: 
            s.add(temp)
            k = temp
        else :
            break
    if check : print("YES")
    else : print("NO")
except ValueError:
    print("Loi gia tri")
```

```python
#10
try: 
    a = list(map(int, input().split()))
    dct = {}
    sum = 0
    check = True
    for i in range(1, len(a)): 
        if dct.get(a[i], 0) == 0:
            sum += a[i]
            dct[a[i]] = 1
        else: 
            check = False
            break
    if check : print(sum)
    else: print("Mang khong hop le")
except ValueError: 
    print("Mang khong hop le")
except IndexError: 
    print("Mang khong dap ung so luong phan tu")
```
