```python
#6
def Rounding(x):
    return (f"{x:.2f}")
class Rectangle: 
    def __init__(self, w,h): 
        self.w = w 
        self.h = h
    def area(self): return self.w * self.h
    def perimeter(self): return 2*(self.w + self.h)
    def scale(self,k):
        self.w *= k 
        self.h *= k
w,h,k = map(int, input("Input: ").split())
t = Rectangle(w,h)
t.scale(k) # truyen tham so k ( he so co da) vao ham scaling
print(t.area())
print(t.perimeter())
```
