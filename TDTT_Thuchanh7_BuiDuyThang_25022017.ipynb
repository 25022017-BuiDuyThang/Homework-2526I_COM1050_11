```python
#1
def binary_search(arr, left, right, val):
    mid = (left+right)//2
    if arr[mid] < val: 
        return (binary_search(arr, mid + 1, right, val))
    elif arr[mid] > val:
        return(binary_search(arr, left, mid + 1, val))
    elif arr[mid] == val:
        return mid
    return -1
 
lst = [int(n) for n in input().split()]
k = int(input())
print(binary_search(lst, 0, len(lst) - 1,k))
                        
```

```python
#2
def count_occurences(arr, val):
    c = 0
    for i in arr: 
        if i == val: 
            c+=1
    return c

lst = [int(n) for n in input().split()]
k = int(input())
print(count_occurences(lst,k))
```

```python
#3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n): 
        if i + 1 >= n:
            break
        for j in range(i+1,n - 1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] =temp
    return arr
lst = [int(n) for n in input().split()]
print(bubble_sort(lst))
```

```python
#8
# xử lý input
s1 = input().strip()
# bỏ [[ ]] rồi tách thành các cặp
s1 = s1[2:-2]              # bỏ [[ và ]]
pairs = s1.split("], [")   # tách từng khoảng

intervals = []
for p in pairs:
    l, r = map(int, p.split(","))
    intervals.append((l, r))

# đọc dòng queries
s2 = input().strip()
s2 = s2[1:-1]              # bỏ [ ]
queries = list(map(int, s2.split(",")))
res = []
for i in queries: 
    temp = 1e10
    for j in intervals: 
       if j[1] >= i and j[0] <= i: 
            temp = min(temp, j[1] - j[0])
    if temp == 1e10: 
        res.append(-1)
    else: 
        res.append(temp)
print(res)
```
