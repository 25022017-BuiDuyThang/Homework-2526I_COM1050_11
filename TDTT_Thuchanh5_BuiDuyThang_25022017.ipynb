```python
#W5A1
def maximum(a,b):
    if a > b: 
        return a
    return b
a,b = map(int, input().split())
print(maximum(a,b))
```

```python
#W5A2
def swap(a,b):
    temp = a
    a = b 
    b = temp
    return a,b 
a,b = map(int, input().split()) 
print(swap(a,b))
```

```python
#W5A3
import math
def isPrime(n):
    for i in range(2,int(math.sqrt(n) + 1)):
        if(n%i == 0) :
            return False 
    return True

n = int(input())
print(isPrime(n))
```

```python
#W5A4
def perfecNum(n):
    temp = 0
    i = 1
    while temp <= n: 
        temp = (2**(i-1)) *(2**i - 1)
        if temp == n: return True
        i += 1

    return False
num = int(input())
print(perfecNum(num))
```

```python
#W5A5
def check(a, num): 
    for i in range(len(a)):
        if a[i] == num: return i + 1
    return -1
arr = list(map(int,input().split()))
k = int(input())
print(check(arr, k))
```

```python
#W5A6
def factorial(num): 
    i = 2
    res = 1 
    while i <= num: 
        res = res * i
        i += 1 
    return res
n = int(input())
print(factorial(n))
```

```python
#W5A8
def toBinary(n):
    lst = [] 
    while n > 0: 
        lst.append(n%2)
        n //= 2
    return lst[::-1] # đảo ngược danh sách
    
a,b = map(int, input().split())
bina = toBinary(a)
binb = toBinary(b)
count = 0
max_len = max(len(bina), len(binb))
bina = [0]*(max_len - len(bina)) + bina # NOTICE
binb = [0]*(max_len - len(binb)) + binb

print(bina)
print(binb)
for i in range(max_len): 
    if(bina[i] != binb[i]): 
        count += 1 
print(count)
    
```

```python
#W5A9
def sumcNum(n): 
    res = 0
    while n > 0: 
        res += n % 10
        n //= 10
    return res 

n = int(input())
print(sumcNum(n))
    
```

```python
#W5A10
def mapping(x): 
    mp = {} # similar map in c++ , its called dictionary in pyth
    # or mp = dict()
    lst = []
    sig = 0 
    for i in x: 
        if i not in mp: 
            mp[i] = sig
            sig += 1
        lst.append(mp[i])
    return lst
    
a,b = map(str, input().split())
aura = mapping(a)
beta = mapping(b)
#print(aura)
#print(beta)

check = True
for i in range(len(aura)):
    if aura[i] != beta[i]:
        check = False
        break
if check: 
    print(True)
else:
    print(False)
```
