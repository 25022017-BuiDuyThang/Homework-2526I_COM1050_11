```python
#W5A1
def maximum(a,b):
    if a > b: 
        return a
    return b
a,b = map(int, input().split())
print(maximum(a,b))
```

```python
#W5A2
def swap(a,b):
    temp = a
    a = b 
    b = temp
    return a,b 
a,b = map(int, input().split()) 
print(swap(a,b))
```

```python
#W5A3
import math
def isPrime(n):
    for i in range(2,int(math.sqrt(n) + 1)):
        if(n%i == 0) :
            return False 
    return True

n = int(input())
print(isPrime(n))
```

```python
#W5A4
def perfecNum(n):
    temp = 0
    i = 1
    while temp <= n: 
        temp = (2**(i-1)) *(2**i - 1)
        if temp == n: return True
        i += 1

    return False
num = int(input())
print(perfecNum(num))
```

```python
#W5A5
def check(a, num): 
    for i in range(len(a)):
        if a[i] == num: return i + 1
    return -1
arr = list(map(int,input().split()))
k = int(input())
print(check(arr, k))
```

```python
#W5A6
def factorial(num): 
    i = 2
    res = 1 
    while i <= num: 
        res = res * i
        i += 1 
    return res
n = int(input())
print(factorial(n))
```

```python
#W5A9
def sumcNum(n): 
    res = 0
    while n > 0: 
        res += n % 10
        n //= 10
    return res 

n = int(input())
print(sumcNum(n))
    
```
