```python
#1
def merge(a,b):
    i = j = 0 
    res = []
    while i < len(a) and j < len(b): 
        if a[i] < b[j]: 
            res.append(a[i])
            i += 1
        else: 
            res.append(b[j])
            j += 1
    res = res + a[i: ]
    res = res + b[j: ]
    return res

a = [int(n) for n in input().split()]
b = [int(n) for n in input().split()]
print(merge(a,b))
```
